```python
import requests
import json
from pytube import YouTube
from youtube_transcript_api import YouTubeTranscriptApi

# 木棉花頻道id
muse_id = 'UCgdwtyqBunlRb-i-7PnCssQ'

# 取得同一個資料夾下的youtube api key
with open('api_key.txt', 'r') as f:
    data = json.load(f)
api_key = data['api_key']

# 取得木棉花頻道的播放清單內容，並記錄在list中
playlist = []

# 不斷讀取下一頁直到結束
next_page_token = ''
while True:
    
    # youtube playlists api 可以取得頻道下的播放清單資訊
    url = 'https://www.googleapis.com/youtube/v3/playlists?'
    param = 'maxResults=50&part=contentDetails,snippet,id,status&channelId='+muse_id+'&key='+api_key
    next_page ='&pageToken='+next_page_token

    response = requests.get(url + param + next_page)
    response = json.loads(response.text)

    for i in response['items']:
        tmp = {}
        tmp['id'] = i['id']
        tmp['title'] = i['snippet']['title']
        tmp['description'] = i['snippet']['description']
        tmp['itemCount'] = i['contentDetails']['itemCount']
        
        # 檢查title 如果是國語或不是線上看就不append
        if (tmp['title'].find('【國語】') == -1) and (tmp['title'].find('線上看') != -1):
            playlist.append(tmp)
        
    if response.get('nextPageToken') is not None:
        next_page_token = response['nextPageToken']
    else:
        break
        
### 在這個階段應該就要去database更新一下最新的class了，並且把已經消失的播放清單enable=0
### select現在有的class

# 取得播放清單中的影片資訊，並記錄在list中
playlist_item = []

count = 0
for i in playlist:
    if i['itemCount'] > 0:
        print(i['title'])
        count += 1
        
        # 不斷讀取下一頁直到結束
        next_page_token = ''
        video_list = []
        while True:
            url = 'https://www.googleapis.com/youtube/v3/playlistItems?'
            param = 'maxResults=50&part=contentDetails,snippet,id,status&playlistId='+i['id']+'&key='+api_key
            next_page ='&pageToken='+next_page_token
            response = requests.get(url + param + next_page)
            response = json.loads(response.text)
            
            for j in response['items']:
                if j['status']['privacyStatus'] == 'public':
                    video_list.append(j['contentDetails']['videoId'])
            
            if response.get('nextPageToken') is not None:
                next_page_token = response['nextPageToken']
            else:
                break
                
        tmp = {}
        tmp['title'] = i['title']
        tmp['id'] = i['id']
        tmp['description'] = i['description']
        tmp['itemCount'] = i['itemCount']
        tmp['videoId'] = video_list
        
        playlist_item.append(tmp)
        
# 總共的影片數目
print(count)

```
