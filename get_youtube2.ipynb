```python
import requests
import json

# 木棉花頻道id
muse_id = 'UCgdwtyqBunlRb-i-7PnCssQ'

# api key
with open('api_key.txt', 'r') as f:
    data = json.load(f)
api_key = data['api_key']

# 取得木棉花頻道的播放清單內容，並記錄在list中
playlist = []

# 不斷讀取下一頁直到結束
next_page_token = ''
while True:
    url = 'https://www.googleapis.com/youtube/v3/playlists?'
    param = 'maxResults=50&part=contentDetails,snippet,id&channelId='+muse_id+'&key='+api_key
    next_page ='&pageToken='+next_page_token

    response = requests.get(url + param + next_page)
    response = json.loads(response.text)

    for i in response['items']:
        tmp = {}
        tmp['id'] = i['id']
        tmp['title'] = i['snippet']['title']
        tmp['description'] = i['snippet']['description']
        tmp['itemCount'] = i['contentDetails']['itemCount']
        playlist.append(tmp)
        
    if response.get('nextPageToken') is not None:
        next_page_token = response['nextPageToken']
    else:
        break
```

```python
playlist_item = []

count = 0
for i in playlist:
    if i['itemCount'] > 40:
        print(i['title'])
        count += 1
        next_page_token = ''
        video_list = []
        while True:
            url = 'https://www.googleapis.com/youtube/v3/playlistItems?'
            param = 'maxResults=50&part=contentDetails,snippet,id&playlistId='+i['id']+'&key='+api_key
            next_page ='&pageToken='+next_page_token
            response = requests.get(url + param + next_page)
            response = json.loads(response.text)
            
            for j in response['items']:
                video_list.append(j['contentDetails']['videoId'])
            
            if response.get('nextPageToken') is not None:
                next_page_token = response['nextPageToken']
            else:
                break
                
        tmp = {}
        tmp['title'] = i['title']
        tmp['id'] = i['id']
        tmp['description'] = i['description']
        tmp['itemCount'] = i['itemCount']
        tmp['videoId'] = video_list
        
        playlist_item.append(tmp)
        
print(count)
```
