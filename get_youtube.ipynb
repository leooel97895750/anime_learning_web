```python
import requests
from bs4 import BeautifulSoup

channel_url = "https://www.youtube.com/@MuseTW/playlists"

playlist_raw_data = requests.get(channel_url)
soup = BeautifulSoup(playlist_raw_data.text, "html.parser")
items = soup.find_all('div',id="items")

print(soup.prettify())
```

```python
import re
import json

ytInitialData = {}

pattern = "var ytInitialData = "
script_text = soup.find_all('script')
for i in script_text:
    if re.findall(r'^' + pattern, i.text):
        # print(i.text[len(pattern):-1])
        ytInitialData = json.loads(i.text[len(pattern):-1])
```
