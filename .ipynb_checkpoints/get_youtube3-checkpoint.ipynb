```python
import requests
import json
from pytube import YouTube
from youtube_transcript_api import YouTubeTranscriptApi

# Êú®Ê£âËä±È†ªÈÅìid
muse_id = 'UCgdwtyqBunlRb-i-7PnCssQ'

# ÂèñÂæóÂêå‰∏ÄÂÄãË≥áÊñôÂ§æ‰∏ãÁöÑyoutube api key
with open('api_key.txt', 'r') as f:
    data = json.load(f)
api_key = data['api_key']

# ÂèñÂæóÊú®Ê£âËä±È†ªÈÅìÁöÑÊí≠ÊîæÊ∏ÖÂñÆÂÖßÂÆπÔºå‰∏¶Ë®òÈåÑÂú®list‰∏≠
playlist = []

# ‰∏çÊñ∑ËÆÄÂèñ‰∏ã‰∏ÄÈ†ÅÁõ¥Âà∞ÁµêÊùü
next_page_token = ''
while True:
    
    # youtube playlists api ÂèØ‰ª•ÂèñÂæóÈ†ªÈÅì‰∏ãÁöÑÊí≠ÊîæÊ∏ÖÂñÆË≥áË®ä
    url = 'https://www.googleapis.com/youtube/v3/playlists?'
    param = 'maxResults=50&part=contentDetails,snippet,id&channelId='+muse_id+'&key='+api_key
    next_page ='&pageToken='+next_page_token

    response = requests.get(url + param + next_page)
    response = json.loads(response.text)

    for i in response['items']:
        tmp = {}
        tmp['id'] = i['id']
        tmp['title'] = i['snippet']['title']
        tmp['description'] = i['snippet']['description']
        tmp['itemCount'] = i['contentDetails']['itemCount']
        
        # Ê™¢Êü•title Â¶ÇÊûúÊòØÂúãË™ûÊàñ‰∏çÊòØÁ∑ö‰∏äÁúãÂ∞±‰∏çappend
        if (tmp['title'].find('„ÄêÂúãË™û„Äë') == -1) and (tmp['title'].find('Á∑ö‰∏äÁúã') != -1):
            playlist.append(tmp)
        
    if response.get('nextPageToken') is not None:
        next_page_token = response['nextPageToken']
    else:
        break
        
### Âú®ÈÄôÂÄãÈöéÊÆµÊáâË©≤Â∞±Ë¶ÅÂéªdatabaseÊõ¥Êñ∞‰∏Ä‰∏ãÊúÄÊñ∞ÁöÑclass‰∫ÜÔºå‰∏¶‰∏îÊääÂ∑≤Á∂ìÊ∂àÂ§±ÁöÑÊí≠ÊîæÊ∏ÖÂñÆenable=0
### selectÁèæÂú®ÊúâÁöÑclass

# ÂèñÂæóÊí≠ÊîæÊ∏ÖÂñÆ‰∏≠ÁöÑÂΩ±ÁâáË≥áË®äÔºå‰∏¶Ë®òÈåÑÂú®list‰∏≠
playlist_item = []

count = 0
for i in playlist:
    if i['itemCount'] > 0:
        print(i['title'])
        count += 1
        
        # ‰∏çÊñ∑ËÆÄÂèñ‰∏ã‰∏ÄÈ†ÅÁõ¥Âà∞ÁµêÊùü
        next_page_token = ''
        video_list = []
        while True:
            url = 'https://www.googleapis.com/youtube/v3/playlistItems?'
            param = 'maxResults=50&part=contentDetails,snippet,id,status&playlistId='+i['id']+'&key='+api_key
            next_page ='&pageToken='+next_page_token
            response = requests.get(url + param + next_page)
            response = json.loads(response.text)
            
            for j in response['items']:
                if j['status']['privacyStatus'] == 'public':
                    video_list.append(j['contentDetails']['videoId'])
            
            if response.get('nextPageToken') is not None:
                next_page_token = response['nextPageToken']
            else:
                break
                
        tmp = {}
        tmp['title'] = i['title']
        tmp['id'] = i['id']
        tmp['description'] = i['description']
        tmp['itemCount'] = i['itemCount']
        tmp['videoId'] = video_list
        
        playlist_item.append(tmp)
        
# Á∏ΩÂÖ±ÁöÑÂΩ±ÁâáÊï∏ÁõÆ
print(count)

```
