## 1. 找出木棉花的播放清單 playlist
* 標題不能是【國語】
* 標題要有線上看
* status要是公開

```python
import requests
import json
from pytube import YouTube
from youtube_transcript_api import YouTubeTranscriptApi

# 木棉花頻道id
muse_id = 'UCgdwtyqBunlRb-i-7PnCssQ'

# 取得同一個資料夾下的youtube api key
with open('api_key.txt', 'r') as f:
    data = json.load(f)
api_key = data['api_key']

# 取得木棉花頻道的播放清單內容，並記錄在list中
playlist = []

# 不斷讀取下一頁直到結束
next_page_token = ''
while True:
    
    # youtube playlists api 可以取得頻道下的播放清單資訊
    url = 'https://www.googleapis.com/youtube/v3/playlists?'
    param = 'maxResults=50&part=contentDetails,snippet,id,status&channelId='+muse_id+'&key='+api_key
    next_page ='&pageToken='+next_page_token

    response = requests.get(url + param + next_page)
    response = json.loads(response.text)
    # print(response)

    for i in response['items']:
        tmp = {}
        tmp['id'] = i['id']
        tmp['title'] = i['snippet']['title']
        tmp['description'] = i['snippet']['description']
        tmp['itemCount'] = i['contentDetails']['itemCount']
        
        # 檢查是不是公開影片
        if i['status']['privacyStatus'] != 'public':
            print(tmp['title'] + ' is not a public playlist')
            continue
            
        # 檢查title有沒有國語
        if tmp['title'].find('【國語】') != -1:
            print(tmp['title'] + ' is 【國語】')
            continue
            
        # 檢查title是不是線上看
        if tmp['title'].find('線上看') == -1:
            print(tmp['title'] + ' maybe is a trailer')
            continue
            
        playlist.append(tmp)
        
    if response.get('nextPageToken') is not None:
        next_page_token = response['nextPageToken']
    else:
        break
```

## 2. 找出每個播放清單裡面的影片 playlist_item
* 檢查影片是不是公開的

```python
playlist_item = []

for i in range(0, len(playlist)):

    if (i in prefer_list) and (playlist[i]['itemCount'] > 0):
        # 不斷讀取下一頁直到結束
        next_page_token = ''
        video_list = []
        while True:
            url = 'https://www.googleapis.com/youtube/v3/playlistItems?'
            param = 'maxResults=50&part=contentDetails,snippet,id,status&playlistId='+playlist[i]['id']+'&key='+api_key
            next_page ='&pageToken='+next_page_token
            response = requests.get(url + param + next_page)
            response = json.loads(response.text)
            
            for j in response['items']:
                
                # 檢查是不是公開的影片
                if j['status']['privacyStatus'] == 'public':
                    video_list.append(j['contentDetails']['videoId'])
                else:
                    print(j['contentDetails']['videoId'] + 'is private video')
            
            if response.get('nextPageToken') is not None:
                next_page_token = response['nextPageToken']
            else:
                break
                
        tmp = {}
        tmp['title'] = playlist[i]['title']
        tmp['id'] = playlist[i]['id']
        tmp['description'] = playlist[i]['description']
        tmp['itemCount'] = playlist[i]['itemCount']
        tmp['videoId'] = video_list
        
        playlist_item.append(tmp)
```

## 擷取日文字幕

```python
# 資料庫的操作先定義好
import pymysql.cursors
import json

with open('password.txt', 'r') as f:
    data = json.load(f)

# Connect to the database
connection = pymysql.connect(host = 'localhost',
                             user = 'root',
                             password = data['password'],
                             database = 'test_db',
                             cursorclass = pymysql.cursors.DictCursor)


# 插入class
def sql_insert_class(type_class, CName, CDes, PlayListID):
    cid = 0
    with connection:
        with connection.cursor() as cursor:

            sql = "INSERT INTO `class` (`type`, CName, CDes, PlayListID) VALUES (%s, %s, %s, %s)"
            cursor.execute(sql, (type_class, CName, CDes, PlayListID))
            cid = cursor.lastrowid

        connection.commit()
    return cid

# 插入object sentence co
def sql_insert_object_sentence_co(cid, srt):
    
    with connection:
        with connection.cursor() as cursor:

            # 插入全部的字幕並建立關聯
            for i in range(0, len(srt)):
                
                sql = "INSERT INTO `object` (`type`) VALUES (%s)"
                cursor.execute(sql, (1,))
                oid = cursor.lastrowid
                
                sql = "INSERT INTO `sentence` (SID, JText, `Start`, `Duration`) VALUES (%s, %s, %s, %s)"
                cursor.execute(sql, (oid, srt[i]['text'], srt[i]['Start'], srt[i]['Duration']))

                sql = "INSERT INTO `co` (CID, OID) VALUES (%s, %s)"
                cursor.execute(sql, (cid, oid))

        connection.commit()
```

```python
# 擷取各影片的字幕
for i in range(0, len(playlist_item)):
    
    print('progress: ' + str(round((i / len(playlist_item)) * 100)) + '%')
    
    # sql insert class
    cid = sql_insert_class(1, playlist_item[i]['title'], playlist_item[i]['description'], playlist_item[i]['id'])
    
    video_list = playlist_item[i]['videoId']
    for j in range(0, len(video_list)):
        
        # 用try except避免意料外的錯誤導致程式終止
        try:
            # 取得自動產生的日文字幕
            srt = YouTubeTranscriptApi.get_transcript(video_list[j], languages=['ja'])
            sql_insert_object_sentence_co(cid, srt)
            
        except:
            print('some thing wrong in ' + video_list[j])
```
